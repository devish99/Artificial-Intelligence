```python
'''
Importing the 'random' module
To randomize the starting point of the Vacuum Cleaner
And to randomize the states of the locations
'''
import random

#Ordered Dictionary to initialize random states of the location
# Kitchen is location A
# #Bedroom is location B
Condition = {'A': '0', 'B': '0'}

Condition['A'] = random.randint(0, 1)
Condition['B'] = random.randint(0, 1)
print (Condition)

#Variable to measure the performance of the Vacuum Cleaner
Score = 0

# place vacuum at random location
Location = random.randint(0, 1)

'''
Kitchen and Bedroom are environments.
Suck and Clean are actions performed by the actuators.
'''

# if vacuum is placed at A
if Location == 0:
    print ("Vacuum is randomly placed in the Kitchen")
    
    # and Location A is Dirty.
    if Condition['A'] == 1:
        print ("Kitchen is Dirty")
        # suck the dirt and mark it clean
        Condition['A'] = 0
        Score += 1    #Score is increased by 1 after cleaning
        print ("Kitchen has been cleaned")
                
        # move to B
        print ("Moving to Bedroom...")
        Score -= 1    #Score is decreased by 1 after changing location
        
        # if B is Dirty
        if Condition['B'] == 1:
            print ("Bedroom is Dirty.")
            # suck and mark clean
            Condition['B'] = 0
            Score += 1
            print ("Bedroom has been Cleaned.")
            
        else:
            print("Bedroom is already clean")
    else:
        print("Kitchen is already clean")
        # move to B
        Score -= 1
        print ("Moving to Bedroom...")
        
        # if B is Dirty
        if Condition['B'] == 1:
            print ("Bedroom is Dirty.")
            # suck and mark clean
            Condition['B'] = 0
            Score += 1
            print ("Bedroom has been Cleaned.")
            
        else:
            print("Bedroom is already clean")

#if Vacuum is placed at B
elif Location == 1:
    print ("Vacuum randomly placed in the Bedroom.")
    # and B is Dirty

    if Condition['B'] == 1:
        print ("Bedroom is Dirty.")
        # suck and mark clean
        Condition['B'] = 0
        Score += 1
        print ("Bedroom has been Cleaned.")
        # move to A
        Score -= 1
        print ("Moving to Kitchen...")
        
        # if A is Dirty
        if Condition['A'] == 1:
            print ("Kitchen is Dirty.")
            # suck and mark clean
            Condition['A'] = 0
            Score += 1
            print ("Kitchen has been Cleaned.")
            
        else:
            print("Kitchen is already clean")
            
    else:
        # if B is already clean move to A
        print("Bedroom is already clean")
        print ("Moving to Kitchen...")
        Score -= 1
        # if A is Dirty
        if Condition['A'] == 1:
            print ("Kithcen is Dirty.")
            # suck and mark clean
            Condition['A'] = 0
            Score += 1
            print ("Kitchen has been Cleaned.")
            
        else:
            print("Kitchen is already clean")
        
# Cleaning complete
print ("Performance Measurement: " + str(Score))
```
